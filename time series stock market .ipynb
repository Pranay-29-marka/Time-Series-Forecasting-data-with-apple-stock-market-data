```python
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
scaler = MinMaxScaler(feature_range=(0,60))
scaler
scaled_data = scaler.fit_transform(df[['open']])
scaled_data
```

```python
#creating sequence data
def create_dataset(df, time_step=60):
    X, Y = [], []
    for i in range(len(df) - time_step):
        X.append(df[i:(i+time_step), 0])   # past 90 values
        Y.append(df[i+time_step, 0])       # next value
    return np.array(X), np.array(Y)
```

```python

lstm_train_size = int(len(scaled_data) * 0.8)
lstm_train_size
```

```python
lstm_train_data = scaled_data[:lstm_train_size]
lstm_train_size
```

```python
lstm_test_data = scaled_data[lstm_train_size:]
lstm_test_data#predicts scaled data or 20%data
```

```python
#taining &testing the data
X_train, y_train = create_dataset(lstm_train_data)
X_test, y_test = create_dataset(lstm_test_data)
```

```python
X_train, y_train#x_train is useful to trained the actual values &y_train will helpul to sequence of prediction values or comes from x_train sequence
```
